### Question 1: Simple Linear Regression 

Let's set up our workspace and use the `Boston` dataset in the `ISLP` library. Print `Boston` to learn more about the dataset.

```python
# Import standard libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import subplots
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Import specific objects
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from statsmodels.stats.outliers_influence \
     import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm

from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
```

```python
# Load the "Boston" dataset using the "load_data" function from the ISLP package
Boston = load_data('Boston')
```

Before we fit and review model outputs, we should visualize our data. Review the code and plot, shown below. Answer the following questions:

_(i)_ What are the `medv` and `dis` variables being plotted? (Hint: review this [link](https://islp.readthedocs.io/en/latest/datasets/Boston.html))
medv: This variable represents the median value of owner-occupied homes in $1000s.
dis: This variable represents the weighted distances to five Boston employment centers.

_(ii)_ What concept ‘defines’ the plotted line?
The concept that defines the plotted line is the linear regression model. In this context, the plotted line represents the predicted values of medv (median home value) based on the linear relationship with dis (distance to employment centers). The line is defined by the coefficients estimated by the linear regression model, including the intercept and slope. The intercept represents the value of medv when dis is zero, and the slope represents the change in medv for a one-unit increase in dis.

```python
# Extract the variables
medv = Boston['medv'].values.reshape(-1, 1)
dis = Boston['dis'].values.reshape(-1, 1)

# Plot data
plt.scatter(dis, medv, label='Data')
plt.xlabel('dis')
plt.ylabel('medv')

# Fit a linear regression model
lm = LinearRegression()
lm.fit(dis, medv)

# Plot the regression line
plt.plot(dis, lm.predict(dis), color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('dis')
plt.ylabel('medv')
plt.legend()

# Show the plot
plt.show()
```

Consider the variables plotted above. In the context of the `Boston` dataset:

_(iii)_ What is the (implied) null hypothesis? What is the (implied) alternative hypothesis?
Null Hypothesis (H0): There is no linear relationship between dis and medv. 
Alternative Hypothesis (H1): There is a linear relationship between dis and medv. 


_(iv)_ Now, let’s fit a simple regression model, using the general syntax `sm.OLS()` and `model.fit()`. As above, use `medv` as the response variable Y, and `dis` as the predictor variable X. (Hint: use the syntax `sm.add_constant` with the appropriate argument).

```python
# Add your code here
# Add a constant to the predictor variable
X = sm.add_constant(dis)
# Fit the OLS model
model = sm.OLS(medv, X).fit()

# Print the summary of the regression model
print(model.summary())
```

### Question 2: Multiple Linear Regression 

We'll continue to use the `Boston` dataset for this question.

_(i)_ Fit a multiple linear regression, with two predictor variables: $X_1$ is `dis`, and $X_2$ is `rm`. As before, keep `medv` as the response variable Y. (Hint: use the syntax `sm.add_constant` with the appropriate arguments).

```python
# Add your code here
# Add a constant to the predictor variables
predictors_with_const = sm.add_constant(Boston[['dis', 'rm']])

# Fit the multiple linear regression model
model_multiple = sm.OLS(medv, predictors_with_const).fit()

# Print the summary of the model
print(model_multiple.summary())
```

_(ii)_ In the context of the `Boston` dataset, state the null and alternative hypotheses.

_(iii)_ Review the model output, using `summary()`. Does it appear that both `dis` and `rm` are predictive of `medv`? How did you determine this?

Review the first generated plot, 'Residuals vs. Fitted'. Which observations are outliers? What impact might outliers have on our model?

```python
# Add your code here
(ii) the null hypothesis (H0):neither dis nor rm has a significant linear relationship with medv. 
The alternative hypothesis (H1): at least one of the predictor variables (dis or rm) has a significant linear relationship with medv
```

```python
(iii)yes,  as coefficients are statistically significant
```

```python
(iv)any observations that deviate significantly from the horizontal line at 0. 
These observations can be considered outliers.Outliers can skew the estimated coefficients, 
leading to biased parameter estimates. They can also increase the variance of the estimates 
and reduce the model's predictive accuracy. Additionally, outliers can influence model assumptions, 
such as the assumption of normally distributed residuals, leading to invalid inferences. 
```

_(v)_ Fit a second model, this time including an interaction between the two predictor variables. Is there an interaction? (Hint: add a variable `x1 * x2` where `x1` and `x2` are the predictor variables). State an interpretation of the interaction, in the context of the `Boston` dataset, in one or two sentences.

```python
# Add your code here
# Create the interaction term
Boston['dis_rm_interaction'] = Boston['dis'] * Boston['rm']

# Fit the model with the interaction term
model_interaction = sm.OLS(medv, sm.add_constant(Boston[['dis', 'rm', 'dis_rm_interaction']])).fit()

# Print the summary of the model
print(model_interaction.summary())
```

the coefficient associated with the interaction term dis_rm_interaction is statistically significant, it indicates that there is an interaction effect between dis and rm
